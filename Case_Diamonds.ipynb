```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pywedge
```

```python
# Importamos datos de Diamantes

df_diam = pd.read_csv('diamonds.csv', engine='python')

df_diam
```

```python
#Por el momento eliminaremos las columnas x,y y z, ya que nos concentraremos en obtener el monto total del robo.
df_diam = df_diam.drop(columns = ["x","y","z"])
df_diam
```

```python
#Renombramos con mayusculas los encabezados
df_diam = df_diam.rename(columns ={'carat':'Carat', 'cut':'Cut', 'color':'Color', 'clarity':'Clarity','depth':'Depth', 'table':'Table', 'price':'Price'})
```

```python
df_diam.head()
```

```python
#Obtenemos informaciÃ³n de cantidad de datos nulos por columna  

df_diam.info()
```

```python
#Interpolamos datos de la columna depth para obtener datos faltantes
df_diam['Depth'] = df_diam['Depth'].interpolate()
df_diam
```

```python
#Corroboramos y confirmamos que las columnas tienen la misma cantidad de datos
df_diam.info()
```

```python
#Limpiamos columnas de caracteres especiales
df_diam['Cut'] = df_diam['Cut'].str.replace('\W', '', regex=True)
df_diam
```

```python
#Realizamos un conteo inicial tomando el tipo de corte y su costo,nos damos cuenta que tendremos que limpiar esta columna ya que contiene caracteres. 
conteo = df_diam.groupby(['Cut'])[['Price']].sum()
conteo
```

```python
#Limpiamos columnas de caracteres especiales
df_diam['Color'] = df_diam['Color'].str.replace('\W', '', regex=True)
df_diam
```

```python
#Realizamos un conteo inicial tomando el tipo de color y su costo,nos damos cuenta que tendremos que limpiar esta columna ya que contiene caracteres. 
conteo = df_diam.groupby(['Color'])[['Price']].sum()
conteo
```

```python
#Limpiamos columnas de caracteres especiales
df_diam['Clarity'] = df_diam['Clarity'].str.replace('\W', '', regex=True)
df_diam
```

```python
#Realizamos un conteo inicial tomando el tipo de color y su costo,nos damos cuenta que tendremos que limpiar esta columna ya que contiene caracteres. 
conteo = df_diam.groupby(['Clarity'])[['Price']].sum()
conteo
```
