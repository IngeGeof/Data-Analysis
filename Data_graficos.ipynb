```python
#Realizaremos los graficos correspondientes que nos brindan info de los diamantes robados

#Importamos Librerias
import pandas as pd
import cufflinks as cf
from IPython.display import display,HTML
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
#Importamos base de datos estandarizada
df_board = pd.read_csv('datafinal.csv')
```

```python
#Revisamos encabezados
df_board.head()
```

```python
#Por el momento eliminaremos dos  columnas, ya que nos concentraremos en obtener otros graficos para identificar insights.
df_board = df_board.drop(columns = ["Unnamed: 0","Table"])
df_board
```

```python
#Revisamos encabezados de columnas
df_board.columns
```

```python
#Exportamos data lista para usarlo en el Dashboard
df_board.to_csv('datagrap.csv', index=False)
```

```python
#GrÃ¡fico 1: Obtenemos la relaciÃ³n entre la caracterÃ­stica Depth y el Precio de los diamantes
plt.scatter(df_board["Price"], df_board["Depth"])
plt.xlabel("Price", color = 'green')
plt.ylabel("Depth", color = 'green')
plt.show()
```

```python
#GrÃ¡fico 1: Obtenemos la relaciÃ³n entre el Precio y los quilates de los diamantes
plt.scatter(df_board["Carat"], df_board["Price"])
plt.xlabel("Carat", color = 'green')
plt.ylabel("Price", color ='green')
plt.show()
```

```python
#Obtenemos la suma de la cantidad de diamantes por corte que se robaron
serie_cut = df_board.Cut.value_counts()

#Graficamos este conteo
serie_cut = serie_cut.sort_values(ascending=True)
serie_cut.plot.barh()
```
